### Traning the model

```python
import chemprop
arguments = [
    '--data_path', '../data_files/Final_Dataset.csv',
    '--dataset_type', 'regression',
    '--save_dir', '../D-MPNN/Final_Dataset',
    '--save_smiles_splits',
    '--epochs', '50',
    '--dropout', '0.1',
    '--activation', 'PReLU',
    '--batch_size', '128',
    '--metric', 'mae',
    '--num_folds', '3',
    '--seed', '42',
    '--cache_cutoff', '15000',
    '--no_features_scaling',
    '--init_lr',' 0.001',
    '--max_lr',' 0.001',
    '--final_lr',' 0.001',
    '--ffn_hidden_size', '300',
    '--hidden_size', '300'

]
args = chemprop.args.TrainArgs().parse_args(arguments)
mean_score, std_score = chemprop.train.cross_validate(args=args, train_func=chemprop.train.run_training)
```
